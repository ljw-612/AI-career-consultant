```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
resume_df = pd.read_json('../data/Entity Recognition in Resumes.json', lines=True)

# drop the extras column
resume_df = resume_df.drop(columns=['extras'])
resume_df
```

```python
annotation = resume_df['annotation'][0]
annotation
def get_skills(annotation):
    skills = []
    for i in annotation:
        try:
            if i['label'][0] == 'Skills':
                skill = i['points'][0]['text']
                skill = skill.strip()
                skills.append(i['points'][0]['text'])
        except:
            return np.nan
        
    skill_str = ', '.join(skills)
    if skill_str:
        return skill_str
    else:
        return np.nan

resume_df['skills'] = resume_df['annotation'].apply(get_skills)
resume_df = resume_df.dropna().reset_index(drop=True)
resume_df
```

```python
import re

sentance_df = resume_df[['content']]
sentance_df['sentences'] = sentance_df.index
# Function to split content into words and create a DataFrame
def split_and_frame(row):
    # Splitting the content into words based on spaces
    words = row['content'].replace('\n', ' ')
    # Use regex to identify and separate punctuation
    # Pattern to identify words and punctuation, treating punctuation as separate tokens
    # 定义匹配单词、指定标点或包含特殊字符（如c++）的正则表达式
    pattern = r"\w+\+\+|\w+|[.:!?;,\"']" 
    # pattern = r"\w+|[^\w\s]"
    split_list = re.findall(pattern, words)
    # Creating a DataFrame with these words and the corresponding sentence number
    temp_df = pd.DataFrame({'sentence #': row['sentences'], 'Word': split_list})
    return temp_df

# Applying the function to each row and concatenating the results
words_df = pd.concat([split_and_frame(row) for index, row in sentance_df.iterrows()], ignore_index=True)

words_df
```

```python
# remove rows only have dots
words_df = words_df[words_df['Word'] != '•']
words_df = words_df[words_df['Word'] != '/']
words_df = words_df[words_df['Word'] != '-']
words_df = words_df[words_df['Word'] != '~']
words_df = words_df[words_df['Word'] != '●']
```

```python
import re

meta_skills = {}
skills_lists = [i.split('\n') for i in resume_df['skills']]
for index, skills_list in enumerate(skills_lists):
    skills_list = [i.split(',') for i in skills_list]
    flattened_list = [item for sublist in skills_list for item in sublist if item]
    flattened_list = [item.split('(')[0] for item in flattened_list]
    flattened_list = [item.rstrip('•*❖◦~. ').strip() for item in flattened_list]
    # remove bullet points markers
    # flattened_list = [re.sub(r'[^\w\s]', '', i) for i in flattened_list]
    flattened_list = [item.lstrip('•*❖◦~ ').strip() for item in flattened_list]
    # remove '' from flattened_list
    flattened_list = [i for i in flattened_list if i]
    meta_skills[index] = flattened_list
# meta_skills
```

```python
import json
with open('../data/meta_skills.json', 'r') as f:
    meta_skills = json.load(f)
# meta_skills = json.loads('../data/meta_skills.json')
# meta_skills
```

```python
meta_skills_begin_words = {}
for key, value in meta_skills.items():
    begin_words = []
    for skill in value:
        begin_words.append(skill.split()[0])
    meta_skills_begin_words[key] = begin_words
    
# meta_skills_begin_words
```

```python
meta_skills_inside_words = {}
for key, value in meta_skills.items():
    inside_words = []
    for skill in value:
        inside_words.extend(skill.split()[1:])
    meta_skills_inside_words[key] = inside_words
# meta_skills_inside_words
```

```python
words_df['Tag'] = 'O'

for row in words_df.iterrows():
    sentence_index = str(row[1]['sentence #'])
    word = row[1]['Word'].rstrip(',.').strip()
    if meta_skills_begin_words[sentence_index] != []:
        if word in meta_skills_begin_words[sentence_index]:
            words_df.at[row[0], 'Tag'] = 'B-Skill'
        elif word in meta_skills_inside_words[sentence_index]:
            words_df.at[row[0], 'Tag'] = 'I-Skill'
    
words_df
```

```python
# for each sentence #, keep the last 40 rows
# words_df = words_df.groupby('sentence #').tail(100)
# if tag=i-skill and the row above is not b-skill, make this row to be O
words_df = words_df.reset_index(drop=True)
for i in range(1, len(words_df)):
    if words_df.loc[i, 'Tag'] == 'I-Skill' and words_df.loc[i-1, 'Tag'] != 'B-Skill':
        words_df.loc[i, 'Tag'] = 'O'
words_df
```

```python
# if the sentence has no B-skill or i-skill, remove the sentence

# 按照'sentence'进行分组，然后检查每个组的所有'Tag'是否都是'O'
# 仅保留那些不全是'O'的组
words_df = words_df.groupby('sentence #').filter(lambda x: not all(x['Tag'] == 'O'))

words_df = words_df.reset_index(drop=True)
words_df
```

```python
words_df['sentence'] = words_df[['sentence #', 'Word', 'Tag']].groupby('sentence #')['Word'].transform(lambda x: ' '.join(x))
words_df['word_labels'] = words_df[['sentence #', 'Word', 'Tag']].groupby('sentence #')['Tag'].transform(lambda x: ','.join(x))
words_df.head()
```

```python
words_df = words_df[["sentence", "word_labels"]].drop_duplicates().reset_index(drop=True)

words_df
```

```python
# if the word_labels is all O, remove the row
for index, row in words_df.iterrows():
    word_labels_list = row['word_labels'].split(',')
    if 'B-Skill' not in word_labels_list and 'I-Skill' not in word_labels_list:
        words_df.drop(index, inplace=True)
words_df = words_df.reset_index(drop=True)
```
